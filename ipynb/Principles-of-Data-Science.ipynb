## **Data Science Workflow**
### Cross Industry Standard Process for Data Mining(**CRISP-DM**):
- **Business Understand**
    - Business objective (business processes, associated costs/pain)
    - Assess situation
    - Define the success criteria
    - Data Science goals
    - Project Plan (assumptions and risk factors, e.g. technical/financial/business/organizational)
- **Data Understanding**
    - Data Collection (original/Secondary sources)
    - Data description (quality, statistics)
    - Data exploration (structured/formation, features)
- **Data Preparation** (more detail in Data Wrangling)
    - Integration (joining, summarisation/aggregation)
    - Selection (sampling, attribute subset)
    - Transformation (log, PCA, normalisation..)
    - Cleaning (missing)
    - Construction (derived attributes)
- **Modeling**
    - Select modelling technique (by Problem type, Output requiremnets)
    - Develop a testing regime
    - Build model (initial paramter settings, model behaviour)
    - Assess model (over fitting, error distribution)
    - Iteratively adjust parameter settings
- **Evaluation**
    - Validate model (domain experts, business usefulness)
    - Review process
    - Determine next steps (Potential, Metrics for success)
- **Deployment**
    - Hosting models on server
    - Building API
    - Scaling infrasture
    - Monitoring performance
    - Schedule periodical retraining
