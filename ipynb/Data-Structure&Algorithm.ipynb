Data structures and algorithms focus on a core issue: how to efficiently organize data and, on this basis, solve computational problems. When the problem scale is very small, almost any implementation method can work. However, as the volume of data increases and demands become more complex, the efficiency differences among various designs will rapidly widen. Data structures determine how data is stored, accessed and modified, while algorithms describe the specific steps to complete tasks under these constraints. Understanding them is not merely about writing faster programs; more importantly, it is about learning to think about problems in an analytical and deductible way.

## Part 1: Algorithm Analysis

Algorithm analysis is the systematic study of an algorithm’s efficiency in terms of resource consumption as a function of the input size $n$. It abstracts away from implementation details such as programming language, hardware, and compiler optimizations, and instead focuses on how an algorithm’s running time grows as the problem size increases. The primary purpose of algorithm analysis is to compare algorithms, evaluate their scalability, and provide theoretical guarantees about performance before actual implementation.

---

### Two Dimensions of Algorithm Analysis

#### Time Complexity $T(n)$

Time complexity describes how the number of **basic operations** performed by an algorithm grows with the input size $n$. A basic operation is usually chosen as the most representative or dominant operation in the algorithm, such as comparisons in sorting or node expansions in graph traversal.

For example, in **Linear Search**, the algorithm checks elements one by one. In the worst case, it examines all $n$ elements, so $T(n)=O(n)$. In **Binary Search**, the search interval is halved at each step, and the number of comparisons grows logarithmically, giving $T(n)=O(\log n)$. These expressions characterize growth rates rather than exact execution time.

#### Space Complexity $S(n)$

Space complexity measures how much **additional memory** an algorithm requires as the input size $n$ grows. This typically refers to auxiliary space, including temporary variables, data structures, and recursion call stacks.

For instance, **Merge Sort** allocates auxiliary arrays to merge subproblems, leading to $S(n)=O(n)$. In contrast, **Insertion Sort** rearranges elements in place and uses only a constant number of extra variables, so $S(n)=O(1)$. Recursive algorithms may also require stack space proportional to their recursion depth.

---

### Asymptotic Analysis Framework

In practice, exact running time is difficult to compute and highly dependent on the execution environment. Therefore, algorithm analysis relies on **asymptotic analysis**, which studies the growth of $T(n)$ and $S(n)$ as $n$ becomes large. The focus is on dominant terms, ignoring constant factors and lower-order terms, since these have negligible impact for sufficiently large $n$.

Three standard performance scenarios are considered:
- **Worst-case analysis**, which provides an upper bound and is the most commonly used in theory and practice.
- **Average-case analysis**, which considers expected performance under an assumed input distribution.
- **Best-case analysis**, which describes the most favorable inputs but is usually of limited practical value.

---

### Asymptotic Notation and Growth Orders

Asymptotic notation provides a formal language to describe algorithm growth rates.

- **Big-O notation**: $T(n)=O(f(n))$ denotes an asymptotic upper bound, meaning that $T(n)$ does not grow faster than $f(n)$.
- **Big-$\Omega$ notation**: $T(n)=\Omega(f(n))$ denotes an asymptotic lower bound.
- **Big-$\Theta$ notation**: $T(n)=\Theta(f(n))$ denotes a tight bound, where $f(n)$ matches the growth rate of $T(n)$ up to constant factors.

Common growth orders, from slower to faster, include:



$$O(1)<O(\log n)<O(n)<O(n\log n)<O(n^2)<O(2^n)<O(n!)$$
As $n$ increases, higher-order growth rates quickly dominate lower-order ones.

---

### Time Complexity Analysis Techniques

#### Loop-Based Analysis

A single loop that runs $n$ times with constant work per iteration has:
$$T(n)=O(n)$$

Nested loops multiply their iteration counts. For example, two nested loops each running $n$ times yield:
$$T(n)=O(n^2)$$

Loops that reduce the problem size multiplicatively, such as repeatedly halving $n$, lead to logarithmic complexity:
$$T(n)=O(\log n)$$

#### Recursive Algorithms and Recurrence Relations

Many algorithms are naturally expressed using recursion, leading to recurrence relations of the form:
$$T(n)=aT(n/b)+f(n)$$
where $a$ is the number of subproblems, $n/b$ is the size of each subproblem, and $f(n)$ represents the cost of dividing and combining.
